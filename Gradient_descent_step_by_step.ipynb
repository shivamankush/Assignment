```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
# lets apply OLS(ordinary least square)
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X,y)
```

```python
# value of slope
reg.coef_
```

```python
# value of intercept
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Let's apply gradient descent assuming slope is constant  m=78.35
# and let's assume the starting value for incercept b=0
ypred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,ypred,color='#00a65a',label='b=0')
plt.legend()
plt.plot()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Let's take learning rate=0.1
lr=0.1

step_size=loss_slope*lr
step_size
```

```python
# calculate the new intercept
b=b-step_size
b
```

```python
ypred1=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X,ypred1,color='#00a65a',label='b={}'.format(b))
plt.plot(X,ypred,color='#A3E407',label='b=0')
plt.legend()
plt.plot()
```

```python
# Iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
# calculate the new intercept
b=b-step_size
b
```

```python
ypred2=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X,ypred2,color='#00a65a',label='b={}'.format(b))
plt.plot(X,ypred1,color='#A3E407',label='b={}'.format(b))
plt.plot(X,ypred,color='#A3E407',label='b=0')
plt.legend()
plt.plot()
```

```python
# Iteration 3
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
# calculate the new intercept
b=b-step_size
b
```

```python
ypred3=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X,ypred2,color='#00a65a',label='b={}'.format(b))
plt.plot(X,ypred2,color='#A3E407',label='b={}'.format(b))
plt.plot(X,ypred1,color='#A3E407',label='b={}'.format(b))
plt.plot(X,ypred,color='#A3E407',label='b=0')
plt.legend()
plt.plot()
```

```python
b=-100
m=78.35
lr=0.1

epochs=10

for i in range(epochs):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    b=b-(lr*loss_slope)

    y_pred=m*X+b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```
